```python
# Selecting only numeric columns
numeric_df = data.select_dtypes(include=['float64', 'int64'])

# Setting up the subplot grid
num_columns = numeric_df.shape[1]
num_rows = (num_columns // 3) + 1

fig, axes = plt.subplots(num_rows, 3, figsize=(20, 5 * num_rows))
axes = axes.flatten()

# Calculate the maximum y-value for the y-axis across all KDE plots
max_y = 0
for column in numeric_df.columns:
    kde = sns.kdeplot(numeric_df[column])
    current_max_y = max([line.get_ydata().max() for line in kde.get_lines()])
    max_y = max(max_y, current_max_y)
    plt.close()  # Close the temporary plot to avoid displaying it

# Plotting KDE for each numerical column with a consistent y-axis scale
for i, column in enumerate(numeric_df.columns):
    sns.kdeplot(numeric_df[column], ax=axes[i], fill=True)
    axes[i].set_title(column)
    axes[i].set_ylim(0, max_y)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Function to plot RMSD
def plot_rmsd(data, y_column, plot_title, y_label, wildtype, save_path):
    plt.figure(figsize=(15, 8))  # Increase the figure size
    # plt.plot(data['mutation_site'], data[y_column], marker='o', markersize=1.5, linestyle='-', color='b', linewidth=0.5, label='Data')
    plt.plot(data['mutation_site'], data[y_column], marker='o', linestyle='None', markersize=1.5, color='b', linewidth=0.5, label='Data')
    
    plt.axhline(y=wildtype, color='r', linestyle='--', label='Wildtype')  # Add a horizontal line for the wildtype value
    plt.text(len(data['mutation_site']) - 1, wildtype, 'Wildtype', color='r', va='bottom', ha='left')

    # Highlight parameters
    highlight_indices = [3682, 2719, 1633]
    highlight_labels = ['Alpha', 'Epsilon', 'Eta/Iota']
    highlight_colors = ['red', 'green', 'orange']

    # Highlight the special points
    for i, idx in enumerate(highlight_indices):
        point_x = data['mutation_site'].iloc[idx]
        point_y = data[y_column].iloc[idx]
        plt.plot(point_x, point_y, marker='o', color=highlight_colors[i], linestyle='None')  # Use specific color
        plt.text(point_x, point_y, highlight_labels[i], fontsize=12, ha='right', weight='bold')  # Add label

    # Set labels and title
    plt.xlabel('Mutational Variants (Sorted)')
    plt.ylabel(f'{y_label}')
    plt.title(f'{plot_title} for 1-step Mutational Variants')

    # Adjust x-axis labels to show only the start of each amino acid
    plt.xticks(ticks=start_indices, labels=amino_acids, rotation=90, fontsize=8)
    
    # Display the plot
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    # Save the plot if save_path is provided
    if save_path:
        plt.savefig(save_path, format='png', dpi=300)
        print(f"Plot saved to {save_path}")

    plt.show()
```

```python
mutations = sorted(data_sorted['mutation'].unique())
sites = np.arange(1, 196)  # Since your site starts at 1 and ends at 195

# Create mesh grids for mutations and sites
mutation_indices, site_indices = np.meshgrid(mutations, sites, indexing='ij')

# Prepare a grid to hold the binding kinetics data
bind_grid = np.zeros_like(mutation_indices, dtype=float)

# Populate the bind grid with your data
for i, mut in enumerate(mutations):
    for j, site in enumerate(sites):
        # Filter data for each mutation and site
        filter_mask = (data_sorted['mutation'] == mut) & (data_sorted['site'] == site)
        if any(filter_mask):
            bind_grid[i, j] = data_sorted.loc[filter_mask, 'bind'].values[0]
```
